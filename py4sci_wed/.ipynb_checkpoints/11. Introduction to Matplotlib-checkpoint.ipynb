## Basic plotting

The main plotting function is called ``plot``:

```python
plt.plot([1,2,3,6,4,2,3,4])
plt.show()
```

In the above example, we only gave a single list, so it will assume the x values are the indices of the list/array.

However, we can instead specify the x values:

```python
plt.plot([3.3, 4.4, 4.5, 6.5], [3., 5., 6., 7.])
plt.show()
```

Matplotlib can take Numpy arrays, so we can do for example:

```python
import numpy as np
x = np.linspace(0., 10., 50)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

The ``plot`` function is actually quite complex, and for example can take arguments specifying the type of point, the color of the line, and the width of the line:

```python
plt.plot(x, y, marker='o', color='green', linewidth=2)
plt.show()
```

The line can be hidden with:

```python
plt.plot(x, y, marker='o', color='green', linewidth=0)
plt.show()
```

If you are interested, you can specify some of these attributes with a special syntax, which you can read up more about in the Matplotlib documentation:

```python
plt.plot(x, y, 'go')  # means green and circles
plt.plot(x+1, y, 'yo') # means yellow and circles
plt.plot(x+2, y, 'b1') # means blue and cross up
plt.plot(x+3, y, 'b2') # means yellow and cross down
plt.show()
```

We start off by loading the ``data/munich_temperatures_average_with_bad_data.txt`` file which we encountered in the Numpy lecture:

```python
# The following code reads in the file and removes bad values
import numpy as np

## Carica i dati 
date, temperature = np.loadtxt('data/munich_temperatures_average_with_bad_data.txt', unpack=True)
## Elimina i "bad values"
keep = np.abs(temperature) < 90 # questo passaggio serve ad eliminare i "bad data" inseriti dal Prof. che sono tutto pari a +90 e -90. Con la funzione abs li toglie tutti in un colpo solo !
## Tieni solo i dati ripuliti dai bad data
date = date[keep]
temperature = temperature[keep]
```

Now that the data has been read in, plot the temperature against time:

```python
# your solution here
x = date
y = temperature
plt.plot(x, y)
plt.show()
```

Next, try plotting the data against the fraction of the year (all years on top of each other). Note that you can use the ``%`` (modulo) operator to find the fractional part of the dates:

```python
# your solution here
date_wo_year = (list(np.float16(sample % int(sample)) for sample in date))

x = date_wo_year
y = temperature
plt.plot(x, y, 'go')
plt.show()
```

## Other types of plots

### Scatter plots

While the ``plot`` function can be used to show scatter plots, it is mainly used for line plots, and the ``scatter`` function is more often used for scatter plots, because it allows more fine control of the markers:

```python
x = np.random.random(100)
y = np.random.random(100)
plt.scatter(x, y)
plt.show()
```

### Histograms

Histograms are easy to plot using the ``hist`` function:

```python
v = np.random.uniform(0., 10., 100)
h = plt.hist(v)  # we do h= to capture the output of the function, but we don't use it
plt.show(h)
```

```python
h = plt.hist(v, range=[-5., 15.], bins=100)
plt.show(h)
```

### Images

You can also show two-dimensional arrays with the ``imshow`` function:

```python
array = np.random.random((64, 64))
plt.imshow(array)
plt.show()
```

And the colormap can be changed:

```python
plt.imshow(array, cmap=plt.cm.gist_heat)
plt.show()
```

## Customizing plots

You can easily customize plots. For example, the following code adds axis labels, and sets the x and y ranges explicitly:

```python
x = np.random.random(100)
y = np.random.random(100)
plt.scatter(x, y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.xlim(0., 1.)
plt.ylim(0., 1.)
plt.show()
```

### Solution 2

```python
# 1.
## Generate 1000 random values following a Normal distribution 
v = np.random.randn(1000)

## Plot a histogram
plt.hist(v)

plt.show()
```

```python
# 2. Change the number of bins to see the Gaussian properly

## Lista di colori predefiniti di Matplotlib (6 tonalit distinte)
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow']

## Numero di bins da testare
bins_list = [10, 20, 50, 100, 200, 400]  # Aggiunto 400 per vedere l'overfitting

## 
plt.hist(v, range=[-4., 4.], bins=10, color='blue', alpha=0.4, label='Distribuzione 1')
plt.hist(v, range=[-4., 4.], bins=20, color='green', alpha=0.4, label='Distribuzione 2')
plt.hist(v, range=[-4., 4.], bins=50, color='red', alpha=0.4, label='Distribuzione 3')
plt.hist(v, range=[-4., 4.], bins=100, color='cyan', alpha=0.4, label='Distribuzione 4')
plt.hist(v, range=[-4., 4.], bins=200, color='magenta', alpha=0.4, label='Distribuzione 5')
plt.hist(v, range=[-4., 4.], bins=400, color='yellow', alpha=0.4, label='Distribuzione 6')

plt.title('Effetto del numero di bins su una Gaussiana')
plt.xlabel('Valore')
plt.ylabel('Conteggio')
plt.legend()

plt.show()

print("1. Si vede che il miglior numero di bin  pari a 20 assomiglia di pi ad una gaussiana.") 
print("\n2. I vari grafici hanno aree diverse. Questo accade perch le curve sono sono normalizzate. \
Osserva la distribuzione 2 chiaramente avendo solo 20 bins  facile che un bin raggiunga i 300 elementi. \
Non pu accadere per una gaussiana con 400 bin e 100 punti che un bin contenga 300 elementi!")
print("\n3. ")
```

```python
# 3. Change the number of bins to see the Gaussian properly

## Calcola manualmente la Gaussiana teorica (senza SciPy)
x = np.linspace(-4, 4, 1000)
mu, sigma = np.mean(v), np.std(v)  # Media e dev.std dei dati
gaussiana = (1/(sigma * np.sqrt(2*np.pi))) * np.exp(-0.5 * ((x - mu)/sigma)**2)

plt.plot(x, gaussiana, 'r-', linewidth=2, label='Gaussiana teorica (NumPy)')
plt.hist(v, range=[-4., 4.], bins=20, color='green', alpha=0.4, label='Distribuzione 2', density=True)


plt.show()

print("in plt.hist() c' un comando che permette di normalizzare la funzione -> density=True")
```

## Exercise 3

The [central limit theorem](http://en.wikipedia.org/wiki/Central_limit_theorem) states that the arithmetic mean of a large number of independent random samples (from any distribution) will approach a normal distribution. You can easily test this with Numpy and Matplotlib:

1. Create an empty array ``total`` with 10000 values (set to 0)
2. Generate 10000 random values uniformly between 0 and 1
3. Add these values to the ``total`` array
4. Repeat steps 2 and 3 10 times
5. Divide ``total`` by 10 to get the mean of the values you added
5. Make a histogram of the values in ``total``

You can also see how the histogram of ``total`` values changes at each step, if you want to see the evolution!

### Solution 3

```python
# 1.
total = np.zeros(10000)
```

```python
# 2.
random_numbers = np.random.rand(10000)  # Array di 10.000 valori in [0, 1)
```

```python
# 3.
total = total + random_numbers
```

```python
# 4.
for i in range(1,9+1):
    random_numbers = np.random.rand(10000)  # Array di 10.000 valori in [0, 1)
    total += random_numbers
```

```python
# 5.
mean_values = total / 10
```

```python
# 6.
plt.hist(mean_values)
plt.show()
```
